```python
import os
import random
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from keras.optimizers import Adam
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
# Define the dataset paths
train_path = '/content/dataset/dataset/train'
test_path = '/content/dataset/dataset/test'
```

```python
# Define the classes in dataset
classes = ['Olu', 'Rose', 'Temple flowers', 'Lotus', 'Daisy', 'Anthurium']
```

```python
print(classes)
```

```python
# Function to load and preprocess the dataset
def load_dataset(dataset_path):
    data = []
    labels = []

    for class_index, class_name in enumerate(classes):
        class_folder_path = os.path.join(dataset_path, class_name)
        for image_name in os.listdir(class_folder_path):
            image_path = os.path.join(class_folder_path, image_name)
            image = Image.open(image_path)
            image = image.resize((224, 224))  # Resize to 224x224 pixels
            image = np.array(image) / 255.0  # Normalize pixel values to [0, 1]
            data.append(image)
            labels.append(class_index)

    return np.array(data), np.array(labels)
```

```python
# Load and preprocess the training dataset
X_train, y_train = load_dataset(train_path)
```

```python
# Convert labels to one-hot encoding
y_train = to_categorical(y_train, num_classes=len(classes))
```

```python
# Split the dataset into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# Define the AlexNet model using Keras Sequential API
model = Sequential()
```

```python
# Layer 1: Convolutional Layer
model.add(Conv2D(filters=96, kernel_size=(11, 11), strides=(4, 4), activation='relu', input_shape=(224, 224, 3)))

# Layer 2: Max Pooling
model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Layer 3: Convolutional Layer
model.add(Conv2D(filters=256, kernel_size=(5, 5), activation='relu'))

# Layer 4: Max Pooling
model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Layer 5: Convolutional Layer
model.add(Conv2D(filters=384, kernel_size=(3, 3), activation='relu'))

# Layer 6: Convolutional Layer
model.add(Conv2D(filters=384, kernel_size=(3, 3), activation='relu'))

# Layer 7: Convolutional Layer
model.add(Conv2D(filters=256, kernel_size=(3, 3), activation='relu'))

# Layer 8: Max Pooling
model.add(MaxPooling2D(pool_size=(3, 3), strides=(2, 2)))

# Layer 9: Flatten
model.add(Flatten())

# Layer 10: Fully Connected Layer
model.add(Dense(units=4096, activation='relu'))

# Layer 11: Dropout (Regularization)
model.add(Dropout(0.5))

# Layer 12: Fully Connected Layer
model.add(Dense(units=4096, activation='relu'))

# Layer 13: Dropout (Regularization)
model.add(Dropout(0.5))

# Layer 14: Output Layer
model.add(Dense(units=len(classes), activation='softmax'))
```

```python
# Compile the model with the updated learning_rate parameter
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
history = model.fit(X_train, y_train, batch_size=64, epochs=70, validation_data=(X_val, y_val))
```

```python
# Calculate final accuracy on the test dataset
X_test, y_test = load_dataset(test_path)
y_test = to_categorical(y_test, num_classes=len(classes))
test_loss, test_accuracy = model.evaluate(X_test, y_test, verbose=1)
print(f'Final Test Loss: {test_loss * 100:.2f}%')
print(f'Final Test Accuracy: {test_accuracy * 100:.2f}%')
```
